결칙치 : occyp_type  
범주형 : gender, car, reality, income_type, edu_type, family_type, house_type, occyp_type  
두 데이터에 대한 전처리가 필요

FLAG_MOBIL 같은경우 모두 1의 데이터를 가지고 있어 없애도 될 듯

```python
from sklearn.preprocessing import LabelEncoder

category = ['gender', 'car', 'reality', 'income_type', 'edu_type',
                        'family_type', 'house_type', 'occyp_type']
label_encoders = {}
for feature in category:
    le = LabelEncoder()
    train_df[feature] = le.fit_transform(train_df[feature])
    label_encoders[feature] = le

# Checking the processed data
train_df.head()
```

```python
for feature, le in label_encoders.items():
    print(f'{feature}:')
    for idx, label in enumerate(le.classes_):
        print(f'    {label} -> {idx}')
```

# 데이터 EDA



### 데이터 분포 확인

```python
import seaborn as sns

#연속형 분포
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
numerical_features = ['income_total', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'family_size', 'begin_month']
for i, feature in enumerate(numerical_features):
    ax = axes.flat[i]
    sns.histplot(train_df[feature], ax=ax, kde=True)
    ax.set_title(f'{feature} Distribution')

plt.tight_layout()
plt.show()

# 범주형 분포
fig, axes = plt.subplots(3, 3, figsize=(18, 15))
categorical_features = ['gender', 'car', 'reality', 'income_type', 'edu_type', 'family_type', 'house_type', 'occyp_type']
for i, feature in enumerate(categorical_features):
    ax = axes.flat[i]
    sns.countplot(x=train_df[feature], ax=ax)
    ax.set_title(f'{feature} Distribution')

plt.tight_layout()
plt.show()
```

### 결측치 분석

```python
# 결측치의 비율을 계산하고 시각화
missing_values = train_df.isnull().mean() * 100
missing_values = missing_values[missing_values > 0]

if not missing_values.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=missing_values.index, y=missing_values)
    plt.xticks(rotation=45)
    plt.title('Missing Values Percentage')
    plt.show()
else:
    print("No missing values in the dataset.")
```

### 타겟 변수 분석

```python
# 타겟 변수 분포 및 타겟에 따른 특징 분포 시각화
sns.countplot(x=train_df['credit'])
plt.title('Target Variable Distribution')
plt.show()

# 타겟 변수에 따른 다른 변수의 분포 확인
for feature in numerical_features:
    sns.boxplot(x=train_df['credit'], y=train_df[feature])
    plt.title(f'{feature} vs Target Variable')
    plt.show()
```

### 상관관계 분석

```python
#FLAG_MOBIL 드랍
train_df = train_df.drop(columns='FLAG_MOBIL')

plt.figure(figsize=(12, 8))
correlation_matrix = train_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

### 이상치 분석

```python
# 박스 플롯을 통해 연속형 변수의 이상치 시각화
fig, axes = plt.subplots(2, 3, figsize=(18, 15))
for i, feature in enumerate(numerical_features):
    ax = axes.flat[i]
    sns.boxplot(train_df[feature], ax=ax)
    ax.set_title(f'{feature} Boxplot')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='credit', data=train_df)
plt.title('Distribution of Target Variable (credit)')
plt.xlabel('Credit Class')
plt.ylabel('Frequency')
plt.show()
```

```python
# 범주형 변수에 대한 타겟 변수에 따른 분포 시각화
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, hue='credit', data=train_df)
    plt.title(f'{feature} vs Credit Class')
    plt.show()
```

```python
# 상관관계 행렬에서 credit과의 상관관계 확인
correlation_matrix = train_df.corr()
credit_corr = correlation_matrix['credit'].sort_values(ascending=False)
print(credit_corr)
```
